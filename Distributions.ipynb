# Distributions


```python
#import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Uniform Distribution

```python
s = np.random.uniform(-1,0,1000)
sns.distplot(s)
plt.show()
```

## Multinomial Distribution

```python
s = np.random.multinomial(20, [1/6.]*6, size=1)
sns.distplot(s)
plt.show()
```

## Normal Distribution

```python
p = np.random.normal(10, 0.1, 1000)
sns.distplot(p)
plt.show()
```
