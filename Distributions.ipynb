# Distributions
There are many distribution curves, each of these distributions has a specific application and use that is appropriate to a particular setting. Some are the well known like the Normal Distribution.

```python
#import packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. np.random.uniform

Generates numbers for uniform distributions

### What is a Uniform Distribution?

A uniform distribution, sometimes also known as a rectangular distribution, is a distribution that has constant probability. The mean and the median coincide.

```python
# Let's use some descriptive statistics to check the mean and the median.
df = np.random.uniform(1,10,1000)
print("Mean = ", np.mean(df))
print("Median = ",np.median(df))
```

```python
# Uniform Distributions are rectangular in shape
s = np.random.uniform(1,10,1000)
plt.title("Uniform Distribution Histogram")
sns.distplot(s)
plt.show()
```

```python
num_of_points = 150

uniform_array_1 = np.random.uniform(0, 1, num_of_points)
uniform_array_2 = np.random.uniform(2, 4, num_of_points)
uniform_array_3 = np.random.uniform(6, 10, num_of_points)

test = sns.scatterplot(uniform_array_1, uniform_array_2)

plt.show(test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

# Load the example iris dataset
diamonds = sns.load_dataset("diamonds")

# Draw a scatter plot while assigning point colors and sizes to different
# variables in the dataset
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
s = sns.scatterplot(x="carat", y="price",
                hue="clarity", size="depth",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=diamonds, ax=ax)
plt.show(s)
```

## 2. np.random.multinomial

Draw samples from a multinomial distribution

### What is a Multinomial Distribution?

```python
s = np.random.multinomial(20, [1/6.]*6, size=1)
sns.distplot(s)
plt.show()
```

## 3.np.random.normal
Draw samples from a Normal Distribution

### what is a Normal Distribution?


Data can be "distributed" (spread out) in different ways. But there are many cases where the data tends to be around a central value with no bias left or right, and it gets close to a "Normal Distribution" AKA the Bell curve. The normal distributions occurs often in nature. For example, it describes the commonly occurring distribution of samples influenced by a large number of tiny, random disturbances, each with its own unique distribution

In a normal distribution: 
- the mean = median = mode
- symmetry about the centre
- 50% right and 50% left of the mean

```python
p = np.random.normal(10, 0.1, 1000)
sns.distplot(p)
plt.show()
```

## 4. numpy.random.triangular

Draw samples from the triangular distribution over the interval [left, right]

### What is the Triangular Distribution?
The triangular distribution is a continuous probability distribution with lower limit a, upper limit b and mode c, where a < b and a ≤ c ≤ b.

```python
z = np.random.triangular(-20, 0, 80, 100000)

plt.hist(z, density=True)

plt.show()
```

## 5. numpy.random.chisquare

Draw samples from a chi-square distribution

### What is a Chisquare Distribution?

```python
p = np.random.chisquare(2,4)

plt.hist(p, density=True)
plt.show()
```
